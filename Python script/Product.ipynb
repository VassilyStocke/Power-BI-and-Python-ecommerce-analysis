```python
#rename the file containing category name in English into fd and import data via pandas read csv 
import pandas as pd
import numpy as np
fd=pd.read_csv('fd.csv')
```

```python
#Merge the dataframe
df= pd.merge(df,fd, on='product_category_name')
```

```python
# replace underscores with spaces and capitalize each word in the column
df['product_category_name'] = df['product_category_name'].str.replace('_', ' ').str.title()
df['product_category_name_english'] = df['product_category_name_english'].str.replace('_', ' ').str.title()
```
