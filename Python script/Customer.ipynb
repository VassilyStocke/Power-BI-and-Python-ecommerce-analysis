```python
#rename the file df and import data via pandas read csv 
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('df.csv')
duplicates = df[df.duplicated(subset=['customer_id'], keep=False)]
duplicates
```

```python
#Drop unnecessary columns 
df.drop(columns=['customer_city','customer_state'],inplace=True)
```

```python
#Check for outlier
import seaborn as sns
sns.boxplot(data=df[['customer_zip_code_prefix']], orient='h')
```

```python
null_rows = df[df.isnull().any(axis=1)]

# print out the null rows
print(null_rows)
```

```python
#Enforce 5 digit for the zip code
df['customer_zip_code_prefix'] = df['customer_zip_code_prefix'].astype(str).str.zfill(5)
```
