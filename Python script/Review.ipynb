# SENTIMENT ANALYSIS PORTUGESE

```python
#Clean portugese comment 

import nltk
import string
nltk.download('punkt')

# Download stopwords from NLTK
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('portuguese'))

# Define function to remove stopwords and punctuation
def clean_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenize text
    tokens = nltk.word_tokenize(text.lower())
    # Remove stopwords
    tokens = [token for token in tokens if token not in stopwords]
    # Join tokens back into string
    text = ' '.join(tokens)
    return text

# Apply clean_text function to 'comment' column in 'df'
df['review_comment_message_clean'] = df['review_comment_message'].apply(clean_text)
```

```python
#install keyword extractor
!pip install yake
import yake
```

```python
#Extract keyword in portugese
kw_extractor = yake.KeywordExtractor(lan='pt', n=1, top=3)

def extract_keywords(comment):
    keywords = kw_extractor.extract_keywords(comment)
    return [kw[0] for kw in keywords]

df['comment_keyword'] = df['review_comment_message_clean'].apply(lambda x: extract_keywords(x))
```

```python
df.head()
```

```python
#Sentiment analysis portugese 1
from textblob import TextBlob
def get_sentiment(comment):
    blob = TextBlob(comment)
    return blob.sentiment.polarity
df['sentiment'] = df['review_comment_message_clean'].apply(lambda x: get_sentiment(x))
```

```python
df['comment_keyword'] = df['comment_keyword'].astype(str)
```

```python
#Sentiment analysis portugese 2
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
def get_sentiment1(comment):
    scores = sia.polarity_scores(comment)
    return scores['compound']
df['sentiment2'] = df['review_comment_message_clean'].apply(lambda x: get_sentiment1(x))
```

# Sentiment analysis in English

```python
#Clean english comment 

import nltk
import string
nltk.download('punkt')

# Download stopwords from NLTK
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))

# Define function to remove stopwords and punctuation
def clean_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenize text
    tokens = nltk.word_tokenize(text.lower())
    # Remove stopwords
    tokens = [token for token in tokens if token not in stopwords]
    # Join tokens back into string
    text = ' '.join(tokens)
    return text

# Apply clean_text function to 'comment' column in 'df'
df['translated_comment_clean'] = df['translated_comment'].apply(clean_text)
```

```python
#Extract keyword in english
kw_extractor = yake.KeywordExtractor(lan='en', n=1, top=3)

def extract_keywords(comment):
    keywords = kw_extractor.extract_keywords(comment)
    return [kw[0] for kw in keywords]

df['comment_keyword_en'] = df['translated_comment_clean'].apply(lambda x: extract_keywords(x))
```

```python
#Get sentiment from text blob 
from textblob import TextBlob
def get_sentiment(comment):
    blob = TextBlob(comment)
    return blob.sentiment.polarity
df['sentiment_en'] = df['translated_comment_clean'].apply(lambda x: get_sentiment(x))
```

```python
#Get sentiment from nltkSA
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
def get_sentiment1(comment):
    scores = sia.polarity_scores(comment)
    return scores['compound']
df['sentiment_en2'] = df['translated_comment_clean'].apply(lambda x: get_sentiment1(x))
```

```python
df.drop(['translated_comment_clean', 'sentiment','sentiment2', 'sentiment_en','review_comment_message_clean'], axis=1, inplace=True)
```
