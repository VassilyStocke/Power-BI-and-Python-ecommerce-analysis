```python
#rename the file df and import data via pandas read csv 
import pandas as pd
import numpy as np
df=pd.read_csv('df.csv')
df
```

```python
#drop the unecessary column
df.drop(columns=['seller_city','seller_state'],inplace=True)
```

```python
#Check for outlier
import seaborn as sns
sns.boxplot(data=df[['seller_zip_code_prefix']], orient='h')
```

```python
null_rows = df[df.isnull().any(axis=1)]

# print out the null rows
print(null_rows)
```

```python
#Check for duplicate
duplicates = df[df.duplicated(subset=['seller_id'], keep=False)]
duplicates
```

```python
#Enforce 5 digit for the zip code prefix
df['seller_zip_code_prefix'] = df['seller_zip_code_prefix'].astype(str).str.zfill(5)
df
```

```python
#Export the document
from google.colab import files
df.to_csv('sellerbecom.csv') 
files.download('sellerbecom.csv')
```
